```python
# Import libraries that I will be using
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
pd.set_option('display.max_columns',None)
```

```python
# Read the csv file
df = pd.read_csv('heart.csv')
df.head()
```

```python
# Sort into categorical and numerical columns
cat = df[['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']]
num = df[['Age','RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak','HeartDisease']]
```

```python
for i in num.columns:
    sns.histplot(num[i])
    plt.title(i)
    plt.show()
```

```python
for i in cat.columns:
    sns.countplot(x=cat[i], order=cat[i].value_counts().index)
    plt.title(i)
    plt.show()
```

```python
for i in df.columns:
    sns.histplot(data=df, x=i, hue='HeartDisease')
    plt.show()
```

```python
# Notes:
# In this dataset, people age 50-60 are more likely to get heart disease
# Males have more heart disease than females
# People with asymptomatic (ASY) chest pain are more likely to get heart disease
# People with Resting blood pressure of 120-140 are more likely to get heart disease
# People with 0 cholesterol has a higher chance of getting heart disease than people with 200-260dl cholesterol
# People who exercise with angina has a higher chance of getting heart disease
# People with a flat ST Slope are more likely to get heart disease
# In this dataset there are 508 people with heart disease and 410 without
# I decided to 'for loop' over every column and plotting it because it is much faster
```
